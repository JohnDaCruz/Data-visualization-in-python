```python
import yfinance as yf
import plotly as ply
import pandas as pd
import seaborn as sns
import plotly.offline as py
import plotly.graph_objs as go
print(ply.__version__, pd.__version__, sns.__version__, yf.__version__)
```

```python
df = yf.download('PETR4.SA', start='2019-11-01', end='2023-01-04', group_by="ticker")
print(df.head())
print(df.tail())
```

```python
sns.set_theme(style='darkgrid')
sns.displot(data = df['Close'].dropna(),kde=True)
```

```python
dados = [go.Scatter(x=df.index, y=df['Close'])]
layout = go.Layout(title='Histórico dos Preços da Ação 2019-11-01 - 2023-01-04  ',yaxis={'title':'Preços'}, xaxis={'title': 'Período'})
fig = go.Figure(data=dados, layout=layout)
py.iplot(fig)
```
